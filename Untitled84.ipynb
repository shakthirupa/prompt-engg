```python
import pandas as pd
import string
import nltk

nltk.download('stopwords')
nltk.download('wordnet')

from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer

df = pd.read_csv("/content/Twitter Sentiments.csv")

# Drop unnecessary columns
df = df.drop(columns=['id', 'label'])

# Assuming text column is named 'tweet'
text_col = 'tweet'

# Lowercase
df[text_col] = df[text_col].str.lower()

# Remove punctuation
df[text_col] = df[text_col].str.translate(
    str.maketrans('', '', string.punctuation)
)

# Tokenization
df['tokens'] = df[text_col].str.split()

# Stopword removal
stop_words = set(stopwords.words('english'))

df['tokens_no_stopwords'] = df['tokens'].apply(
    lambda words: [w for w in words if w not in stop_words]
)

# Stemming
stemmer = PorterStemmer()

df['stemmed'] = df['tokens_no_stopwords'].apply(
    lambda words: [stemmer.stem(w) for w in words]
)

# Lemmatization
lemmatizer = WordNetLemmatizer()

df['lemmatized'] = df['tokens_no_stopwords'].apply(
    lambda words: [lemmatizer.lemmatize(w) for w in words]
)

print(df.head())
```
